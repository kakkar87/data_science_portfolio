## Data Visualizarion Concepts

In this section we will cover the data visualization using seaborn python library and we will also plot different tips of plots from sample data.

### Distribution Plots

These plots are used for examining univariate and bivariate distributions. Before analyze or model data we need to understand how the variables(features) are distributed. We can get the folloing answers from the distribution plots.
1. What range do the observations cover?
2. What is their central tendency?
3. Skewness: Are they heavily skewed in one direction?
4. Is there evidence for bimodality in the distribution? Bimodality refers to the two local maximums.
5. Are there significant outliers? (Scatter Plots)

#### The axes-level functions like histplot(), kdeplot(), ecdfplot(), and rugplot() are gropued with displot(), jointplot(), and pairplot() functions.

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

We have taken 'Car Price Analysis' dataset from Kaggle.

```python
cars = pd.read_csv('Car Price Analysis.csv')
```

```python
cars.head()
```

```python
 sns.displot(cars['price'], kde=True, bins=35)
```

```python
sns.jointplot(x='carlength', y='carwidth', data=cars, kind='reg')
```

```python
sns.jointplot(x='carlength', y='carwidth', data=cars, kind='kde')
```

```python
sns.pairplot(cars, hue='Affordable') #hue take categorical data 
```

```python
sns.kdeplot(data=cars, x='price')
sns.rugplot(data=cars, x='price')
```

```python
sns.scatterplot(data=cars, x='carlength', y='carwidth', hue='Affordable')
sns.rugplot(data=cars, x='carlength', y='carwidth', hue='Affordable')
```
